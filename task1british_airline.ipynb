### Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
import time
import random

base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 35  # Number of pages to scrape
page_size = 100  # Number of reviews per page

reviews = []

for i in range(1, pages + 1):
    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page with SSL verification turned off
    try:
        response = requests.get(url, verify=False)
        response.raise_for_status()  # Will raise an HTTPError if the request was unsuccessful
        # Explicitly close the connection after the request
        response.close()  
    except requests.exceptions.RequestException as e:
        print(f"Error occurred while scraping page {i}: {e}")
        continue  # Skip this page and continue with the next one

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')

    # Find all review content
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())

    print(f"   ---> {len(reviews)} total reviews")

    # Add a random delay between requests to prevent getting blocked
    time.sleep(random.uniform(5, 10))  # Sleep between 5 and 10 seconds

# Optionally print the reviews to see the output
print(reviews[:5])  # Print the first 5 reviews as a sample
```

 The next thing that you should do is clean this data to remove any unnecessary text from each of the rows. For example, " Trip Verified" can be removed from each row if it exists, as it's not relevant to what we want to investigate.

```python
import pandas as pd
import re

# Load the CSV file containing reviews
df = pd.read_csv('Downloads/BA_reviews.csv')

# Function to clean the review text
def clean_review_text(review_text):
    # Remove unwanted text patterns like " Trip Verified"
    review_text = re.sub(r" Trip Verified", "", review_text)
    review_text = re.sub(r"[^a-zA-Z0-9\s.,!?-]", "", review_text)  # Remove any special characters
    review_text = review_text.strip()  # Remove any leading/trailing whitespace
    return review_text

# Apply the cleaning function to the 'Review' column
df['Review'] = df['Review'].apply(clean_review_text)

# Optionally, print the first few rows to check the cleaned data
print(df.head())

# Save the cleaned reviews to a new CSV file
df.to_csv("Downloads/BA_reviews_cleaned.csv", index=False)
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt

# Load your cleaned reviews CSV
df = pd.read_csv("Downloads/BA_reviews_cleaned.csv")

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get compound sentiment score
def get_vader_sentiment(review_text):
    if pd.isna(review_text):
        return 0  # Handle NaN
    return analyzer.polarity_scores(str(review_text))['compound']

# Apply sentiment function
df['Sentiment'] = df['Review'].apply(get_vader_sentiment)

# Print sentiment stats
print("Sentiment Score Summary:")
print(df['Sentiment'].describe())

# Label the sentiment based on compound score
def label_sentiment(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment_Label'] = df['Sentiment'].apply(label_sentiment)

# Print sentiment label counts
print("\nSentiment Label Counts:")
print(df['Sentiment_Label'].value_counts())

# Optional: Plot sentiment label distribution
sns.countplot(data=df, x='Sentiment_Label', palette='Set2')
plt.title("Sentiment Distribution of British Airways Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords
import re

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load the cleaned reviews CSV
df = pd.read_csv("Downloads/BA_reviews_cleaned.csv")

# Drop NaN values
df = df.dropna(subset=['Review'])

# Text preprocessing
def preprocess(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # Remove non-letter characters
    text = text.lower()
    words = text.split()
    words = [w for w in words if w not in stop_words and len(w) > 2]
    return ' '.join(words)

df['Processed_Review'] = df['Review'].apply(preprocess)

# Vectorize the text data
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df['Processed_Review'])

# Fit the LDA model
lda = LatentDirichletAllocation(n_components=5, random_state=42)  # You can change n_components to adjust topics
lda.fit(doc_term_matrix)

# Display the top words per topic
def print_topics(model, vectorizer, top_n=10):
    for idx, topic in enumerate(model.components_):
        print(f"\n Topic {idx+1}:")
        print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-top_n:]])

print_topics(lda, vectorizer)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import pandas as pd

# Load the cleaned data
df = pd.read_csv("Downloads/BA_reviews_cleaned.csv")

# Drop NaNs and ensure string format
df['Review'] = df['Review'].fillna('').astype(str)

# Use CountVectorizer for LDA
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Review'])

# Fit LDA model for 10 topics
lda = LatentDirichletAllocation(n_components=10, random_state=42)
lda.fit(X)

# Get feature names and top words for each topic
feature_names = vectorizer.get_feature_names_out()
n_top_words = 10
topics = []
for idx, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
    topics.append((f"Topic {idx+1}", top_words))

topics
```

```python
# Assign the most likely topic to each review
def assign_topic(review):
    # Get the topic distribution for the review
    topic_distribution = lda.transform(vectorizer.transform([review]))
    
    # Get the index of the most likely topic
    return topic_distribution.argmax()

# Assign topics to reviews
df['Topic'] = df['Review'].apply(assign_topic)

# Display the first few rows to see the topic assignments
print(df[['Review', 'Topic']].head())
```
